```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
samp = 10**6
mu_S = 5
std_S = 2
S = np.random.normal(mu_S, std_S, (samp,1))


bins = 15
cm = 1/2.54  # centimeters in inches
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 9
fig1, ax1 = plt.subplots(1, figsize=(6*cm, 6*cm), sharey='row', dpi=80, facecolor='w', edgecolor='k')
plt.subplots_adjust(left=0.22, right=.98, top=0.98, bottom=0.1)
ax1.hist(S,bins)
```

```python
S_norm = (S-mu_S)/std_S
```

```python
fig2, ax2 = plt.subplots(1, figsize=(6*cm, 6*cm), sharey='row', dpi=80, facecolor='w', edgecolor='k')
plt.subplots_adjust(left=0.22, right=.98, top=0.98, bottom=0.1)
ax2.hist(S_norm,bins)
```
