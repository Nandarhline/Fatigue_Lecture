```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
samp = 10**4
mu_F = 40000
cov_F = 0.05
F = np.random.normal(mu_F, mu_F*cov_F, (samp,1))

mu_A = 200
cov_A = 0.04
A = np.random.normal(mu_A, mu_A*cov_A, (samp,1))

mu_Sy = 235
cov_Sy = 0.1
Sy = np.random.normal(mu_Sy, mu_Sy*cov_Sy, (samp,1))

bins = 15
cm = 1/2.54  # centimeters in inches
fig1, ax1 = plt.subplots(1, figsize=(6*cm, 6*cm), sharey='row', dpi=80, facecolor='w', edgecolor='k')
plt.subplots_adjust(left=0.22, right=.98, top=0.98, bottom=0.1)
ax1.hist(F/A,bins, label='S')
ax1.hist(Sy,bins, label='S_y')
ax1.legend(loc = 'upper left')
ax1.set_ylim([0, 2600])
```

```python
PF = sum(Sy<F/A)/samp
beta = stats.norm.isf(PF)
PF, beta 
```
