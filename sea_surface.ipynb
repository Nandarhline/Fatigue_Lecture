```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = 22/7
t = np.arange(500)*0.1

## Regular Wave Equation: y = A sin (wt+theta)
y1 = 0.5* np.sin(3*t) 
y2 = 1.2* np.sin(0.3*t) 
y3 = 0.6* np.cos(2*t)
y4 = 0.3* np.sin(t+2*pi)
y5 = 0.05* np.cos(5*t+3*pi)
```

```python
# superposition of 5 regular waves
Y = y1+y2+y3+y4+y5

fig1, ax1 = plt.subplots(6,1, figsize=(17, 8), sharex='all', dpi=80, facecolor='w', edgecolor='k')
plt.subplots_adjust(left=0.08, right=.9, top=0.98, bottom=0.1, hspace = 0.12, wspace=0.15)
ax1[0].plot(y1)
ax1[1].plot(y2)
ax1[2].plot(y3)
ax1[3].plot(y4)
ax1[4].plot(y5)
ax1[5].plot(Y)
ax1[5].set_ylim([-2.5, 2.5])
ax1[5].set_xlabel('time(s)')
```
